```python
import yfinance as yf
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
#stock data
symbol = "^SPX"
data = yf.download(symbol, start="2023-01-01", interval="1d")["Close"]
S0 = data[-1]
#historical volatility
log_returns = np.log(data/data.shift(1))
HV = log_returns.std() * np.sqrt(252)
#r 
r = yf.download("^IRX", start="2022-01-01", interval="1d")["Close"]
r = r[-1]/100
S0, HV, r
```

```python
# option data
symbol = "^SPX"
ticker = yf.Ticker(symbol)
ticker.options
exp_date = ticker.options[22]
option_chain = ticker.option_chain(exp_date)
calls = option_chain.calls
calls
# Filter out SPXW options
calls = calls[~calls['contractSymbol'].str.contains('SPXW')]

# Now, 'calls' DataFrame contains only non-SPXW options
calls
```

```python
savedata = calls.to_csv("SPXCall201023Datigrezzi.csv")
```

```python
calls = calls[calls["openInterest"] > 500]
calls = calls[calls["volume"] > 100]
calls
```

```python
savedata = calls.to_csv("SPXCall201023Datifiltrati.csv")
```

```python
S0 = S0 
sigma = HV 
T = 25 / 252  
r = r 
strike_prices = calls["strike"]

# Black-Scholes call option price
def calculate_option_price(K):
    d1 = (np.log(S0 / K) + (r + (sigma ** 2) / 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    option_price = S0 * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)
    return option_price

calls["OptionPrice"] = round(strike_prices.apply(calculate_option_price),3)


calls.dropna(inplace=True)
calls
```

```python
savedata = calls.to_csv("SPXCall201023Daticonprezzo.csv")
```
